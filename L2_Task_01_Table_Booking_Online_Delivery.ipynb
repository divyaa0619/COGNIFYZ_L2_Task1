```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Dataset (1).csv')
df.head()
df[['Has Table booking', 'Has Online delivery']].head()
table_booking_percentage = (
    df['Has Table booking'].value_counts(normalize=True) * 100
)

print(table_booking_percentage)
online_delivery_percentage = (
    df['Has Online delivery'].value_counts(normalize=True) * 100
)

print(online_delivery_percentage)
avg_rating = df.groupby('Has Table booking')['Aggregate rating'].mean()

print(avg_rating)
sns.barplot(
    x=avg_rating.index,
    y=avg_rating.values
)

plt.title("Average Rating by Table Booking")
plt.ylabel("Average Rating")
plt.show()
sns.countplot(
    x='Price range',
    hue='Has Online delivery',
    data=df
)

plt.title("Online Delivery Across Price Ranges")
plt.show()
```
